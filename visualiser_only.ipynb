```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

# CSV 전처리 함수
def read_and_preprocess_csv(csv_path):
    if not os.path.exists(csv_path):
        print(f"[Error] 파일 없음: {csv_path}")
        return None

    df = pd.read_csv(csv_path, dtype=str)

    for col in df.columns:
        df[col] = df[col].str.strip().replace('', np.nan)
        df[col] = pd.to_numeric(df[col], errors='coerce')

    df.dropna(axis=0, inplace=True)
    return df

# 시계열 그래프 개선

def plot_time_series(df, epi_col='epi', x_col='t', y_col='cgm', title='혈당(CGM) 시계열 변화', max_epi=5):
    episodes = df[epi_col].unique()[:max_epi]

    plt.figure(figsize=(12, 6))

    for epi in episodes:
        epi_df = df[df[epi_col] == epi]
        plt.plot(epi_df[x_col], epi_df[y_col], label=f'Episode {int(epi)}')

    plt.title(title, fontsize=16)
    plt.xlabel('Time step (t)', fontsize=14)
    plt.ylabel('혈당 (CGM, mg/dL)', fontsize=14)
    plt.legend()
    plt.tight_layout()
    plt.show()

# 히스토그램 개선
def plot_histogram(df, col='ins', title='인슐린 투여량(ins) 분포', bins=30):
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], bins=bins, kde=True, color='skyblue')
    plt.title(title, fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('빈도 (Frequency)', fontsize=14)
    plt.tight_layout()
    plt.show()

# 에피소드별 평균 성능 개선

def plot_episode_metric(df, epi_col='epi', metric_col='aBGP_rmse', title='에피소드별 평균 aBGP_rmse (훈련 결과)'):
    epi_metric = df.groupby(epi_col)[metric_col].mean()

    plt.figure(figsize=(12, 6))
    sns.barplot(x=epi_metric.index, y=epi_metric.values, palette='Blues_d')

    plt.title(title, fontsize=16)
    plt.xlabel('Episode', fontsize=14)
    plt.ylabel(f'{metric_col} 평균값', fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":

    csv_paths = {
        "logs_train": "/home/eoeldroal/G2P2C/results/test/training/data/logs_worker_0.csv",
        "summary_train": "/home/eoeldroal/G2P2C/results/test/training/data/training_episode_summary_0.csv",
        "logs_test": "/home/eoeldroal/G2P2C/results/test/testing/data/logs_worker_5000.csv",
        "summary_test": "/home/eoeldroal/G2P2C/results/test/testing/data/testing_episode_summary_5000.csv"
    }

    # Training Logs
    train_logs = read_and_preprocess_csv(csv_paths["logs_train"])
    if train_logs is not None:
        plot_time_series(train_logs, title='훈련 시 혈당(CGM) 변화 (초기 5개 에피소드)')
        plot_histogram(train_logs, col='ins', title='훈련 시 인슐린 투여량 분포(insulin)')

    # Training Episode Summary
    train_summary = read_and_preprocess_csv(csv_paths["summary_train"])
    if train_summary is not None:
        plot_episode_metric(train_summary, metric_col='aBGP_rmse', title='훈련 에피소드별 aBGP RMSE 평균값')

    # Testing Logs
    test_logs = read_and_preprocess_csv(csv_paths["logs_test"])
    if test_logs is not None:
        plot_time_series(test_logs, title='테스트 시 혈당(CGM) 변화 (초기 5개 에피소드)')
        plot_histogram(test_logs, col='rew', title='테스트 시 보상(reward) 분포')

    # Testing Episode Summary
    test_summary = read_and_preprocess_csv(csv_paths["summary_test"])
    if test_summary is not None:
        plot_episode_metric(test_summary, metric_col='cBGP_rmse', title='테스트 에피소드별 cBGP RMSE 평균값')
```

```python
def plot_episode_avg_cgm(df, epi_col='epi', cgm_col='cgm', title='Average CGM per Episode', num_episodes=5):
    """
    에피소드별 평균 CGM 또는 RMSE를 시각화.
    전체 에피소드 중에서 num_episodes만큼 간격을 둬서 고르게 선택하고, 반드시 마지막 에피소드를 포함하도록 구성함.
    """
    if epi_col not in df.columns or cgm_col not in df.columns:
        print(f"해당 DF에 '{epi_col}' 또는 '{cgm_col}' 칼럼이 없습니다. 그래프를 그릴 수 없습니다.")
        return

    unique_episodes = sorted(df[epi_col].unique())
    total_epi = len(unique_episodes)

    # num_episodes 개수에 맞춰서 간격을 설정
    if total_epi <= num_episodes:
        selected_epi = unique_episodes  # 모든 에피소드 선택
    else:
        step = max(total_epi // (num_episodes - 1), 1)
        selected_epi = unique_episodes[::step]

        # 반드시 마지막 에피소드를 포함하도록 함
        if unique_episodes[-1] not in selected_epi:
            selected_epi.append(unique_episodes[-1])

    # 선택한 에피소드로 DataFrame 필터링
    df_selected = df[df[epi_col].isin(selected_epi)]

    # 에피소드별 평균값 계산
    grouped = df_selected.groupby(epi_col)[cgm_col].mean()

    plt.figure(figsize=(10, 6))
    grouped.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(title, fontsize=14)
    plt.xlabel('Episode', fontsize=12)
    plt.ylabel(f"Mean {cgm_col}", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
if __name__ == "__main__":
    csv_paths = {
        "summary_train": "/home/eoeldroal/G2P2C/results/test/training/data/training_episode_summary_0.csv",
        "summary_test": "/home/eoeldroal/G2P2C/results/test/testing/data/testing_episode_summary_5000.csv"
    }

    train_summary = read_and_preprocess_csv(csv_paths["summary_train"])
    if train_summary is not None and not train_summary.empty:
        plot_episode_avg_cgm(
            train_summary, 
            epi_col='epi', 
            cgm_col='aBGP_rmse',
            title='Average aBGP_rmse per Episode (Training)',
            num_episodes=10  # 예: 10개의 에피소드를 고르게 간격을 두어 선택
        )

    test_summary = read_and_preprocess_csv(csv_paths["summary_test"])
    if test_summary is not None and not test_summary.empty:
        plot_episode_avg_cgm(
            test_summary, 
            epi_col='epi', 
            cgm_col='cBGP_rmse',
            title='Average cBGP_rmse per Episode (Testing)',
            num_episodes=10  # 예: 10개의 에피소드를 고르게 간격을 두어 선택
        )
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-darkgrid')

def read_and_preprocess_csv(csv_path):
    df = pd.read_csv(csv_path, dtype=str)
    df = df.apply(lambda x: x.str.strip()).replace('', np.nan).apply(pd.to_numeric, errors='coerce')
    df.dropna(inplace=True)
    return df

def plot_time_series(df, x_col='t', y_col='cgm', group_col='epi', title='CGM Time-Series'):
    plt.figure(figsize=(12, 6))
    for epi in sorted(df[group_col].unique()):
        episode_df = df[df[group_col] == epi]
        plt.plot(episode_df[x_col], episode_df[y_col], alpha=0.3, label=f'Epi {int(epi)}')
    plt.title(title, fontsize=15)
    plt.xlabel('Time Step', fontsize=13)
    plt.ylabel(y_col, fontsize=13)
    plt.show()

def plot_histogram(df, col, title, bins=30):
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], bins=bins, kde=True, color='skyblue')
    plt.title(title, fontsize=15)
    plt.xlabel(col, fontsize=13)
    plt.ylabel("Frequency", fontsize=13)
    plt.show()

def plot_episode_metric(df, epi_col='epi', metric_col='reward', title='Episode-wise Reward'):
    grouped = df.groupby(epi_col)[metric_col].mean()
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=grouped.index, y=grouped.values, marker='o')
    plt.title(title, fontsize=15)
    plt.xlabel('Episode', fontsize=13)
    plt.ylabel(f'Mean {metric_col}', fontsize=13)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def plot_scatter(df, x_col='cgm', y_col='ins', title='CGM vs. Insulin'):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=x_col, y=y_col, alpha=0.3)
    plt.title(title, fontsize=15)
    plt.xlabel(x_col, fontsize=13)
    plt.ylabel(y_col, fontsize=13)
    plt.show()

def plot_boxplot(df, cols, title='Boxplot Analysis'):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df[cols])
    plt.title(title, fontsize=15)
    plt.ylabel('Values', fontsize=13)
    plt.xticks(rotation=45)
    plt.show()

def plot_heatmap(df, title='Feature Correlation Heatmap'):
    plt.figure(figsize=(10, 8))
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title(title, fontsize=15)
    plt.show()

if __name__ == "__main__":
    base_path = "/home/eoeldroal/G2P2C/results/test/"
    csv_paths = {
        "train_logs": base_path + "training/data/logs_worker_0.csv",
        "train_summary": base_path + "training/data/training_episode_summary_0.csv",
        "test_logs": base_path + "testing/data/logs_worker_5000.csv",
        "test_summary": base_path + "testing/data/testing_episode_summary_5000.csv"
    }

    train_logs = read_and_preprocess_csv(csv_paths["train_logs"])
    train_summary = read_and_preprocess_csv(csv_paths["train_summary"])
    test_logs = read_and_preprocess_csv(csv_paths["test_logs"])
    test_summary = read_and_preprocess_csv(csv_paths["test_summary"])

    # 1️⃣ 시계열 분석 (Training)
    plot_time_series(train_logs, y_col='cgm', title='Training CGM Time-Series')

    # 2️⃣ 히스토그램 (Training 인슐린 투여량)
    plot_histogram(train_logs, col='ins', title='Training Insulin Distribution')

    # 3️⃣ 에피소드별 RMSE (Training)
    plot_episode_metric(train_summary, metric_col='aBGP_rmse', title='Training aBGP RMSE per Episode')

    # 4️⃣ Scatter: 혈당 vs. 인슐린 (Training)
    plot_scatter(train_logs, x_col='cgm', y_col='ins', title='Training CGM vs. Insulin')

    # 5️⃣ Boxplot (Training Logs)
    plot_boxplot(train_logs, cols=['cgm', 'ins', 'rew'], title='Training Logs Boxplot Analysis')

    # 6️⃣ Correlation Heatmap (Training)
    plot_heatmap(train_logs[['cgm', 'meal', 'ins', 'rew', 'rl_ins', 'mu', 'sigma', 'prob', 'state_val']])

    # 7️⃣ 시계열 분석 (Testing)
    plot_time_series(test_logs, y_col='cgm', title='Testing CGM Time-Series')

    # 8️⃣ 히스토그램 (Testing 보상 분포)
    plot_histogram(test_logs, col='rew', title='Testing Reward Distribution')

    # 9️⃣ 에피소드별 RMSE (Testing)
    plot_episode_metric(test_summary, metric_col='cBGP_rmse', title='Testing cBGP RMSE per Episode')

    # 🔟 Scatter: 혈당 vs. 보상 (Testing)
    plot_scatter(test_logs, x_col='cgm', y_col='rew', title='Testing CGM vs. Reward')

    # ⭕ Boxplot (Testing Logs)
    plot_boxplot(test_logs, cols=['cgm', 'ins', 'rew'], title='Testing Logs Boxplot Analysis')

    # ⭕ Correlation Heatmap (Testing)
    plot_heatmap(test_logs[['cgm', 'meal', 'ins', 'rew', 'rl_ins', 'mu', 'sigma', 'prob', 'state_val']])
```
